```python
#defining function that will calculate the average word vector for each tweet
def get_avg_vector_for_tweet(tweet_tokens,model):
    vectors=[model.wv[word] for word in tweet_tokens if word in model.wv]
    if len(vectors)>0:
        return np.mean(vectors,axis=0)
    else:
        return np.zeros(model.vector_size)
    
```

```python
train_data["word2vec_features"]=train_data["lemmatized_text"].apply(lambda x:get_avg_vector_for_tweet(x,word2vec_model))
```

```python
train_data.head()
#we have extracted word vectors from the train data set
```

```python
#Converting keywords and locations to word vectors as well using the same model (no need to find mean vector as both keyword and location is 1 word each)
#defining function that will calculate the word vector for each location/keyword
def get_word_vector(input_word,model):
    vector=[model.wv[input_word] if input_word in model.wv else np.zeros(model.vector_size)]
    return vector
```

```python
#Transforming location into vectors:
train_data["word2vec_location"]=train_data["location"].apply(lambda x:get_word_vector(x,word2vec_model))
```

```python
#Transforming keywords into vectors:
train_data["word2vec_keyword"]=train_data["keyword"].apply(lambda x:get_word_vector(x,word2vec_model))
```

```python
#Next, we have to flatten the X variables into a single array as required in scikit-learn
def combine_features(df):
    wv_features = np.stack(df["word2vec_features"].values)
    print("Shape of wv_features:", wv_features.shape)
    
    wv_location = np.stack(df["word2vec_location"].values)
    wv_location=np.squeeze(wv_location)                      
    print("Shape of wv_location:", wv_location.shape)
    
    wv_keyword = np.stack(df["word2vec_keyword"].values)
    wv_keyword=np.squeeze(wv_keyword)   
    print("Shape of wv_keyword:", wv_keyword.shape)
    
    tl = df['tweet_length'].values.reshape(-1, 1)
    print("Shape of tweet_length:", tl.shape)
    
    combined_features = np.concatenate([wv_features, wv_location, wv_keyword, tl], axis=1)
    print("Shape of combined_features:", combined_features.shape)    
    return combined_features
```
